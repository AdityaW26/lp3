y = (x+3)^2

```python
curr_x = 2 # algo begins at x = 2
rate = 0.01 # learning rate
precision = 0.000001 # algo stop
previous_step_size = 1
iters = 0 # counts iterations
max_iters = 10000 # max iterations
df = lambda x : 2*(x+3) # gradient
```

```python
while previous_step_size>precision and iters<max_iters:
    prev_x = curr_x
    curr_x = curr_x - rate*df(prev_x)
    previous_step_size = abs(prev_x-curr_x)
    iters += 1
    print("Iteration no. {0} | Value of x is {1}".format(iters,curr_x))
print("Local minima is found at x = ",curr_x)
```
