```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("emails.csv")
df
```

```python
for i in range(df.shape[0]):
    df['Email No.'][i] = df['Email No.'][i].split(" ")[1]
df
```

```python
cols = list(df.columns)
cols.pop()
cols.pop(0)
len(cols)
```

```python
from wordcloud import WordCloud
wc = WordCloud(background_color="black",width=800,height=600).generate(' '.join(i for i in cols))

plt.figure(figsize=(20,10))
plt.imshow(wc,interpolation='bilinear')
plt.axis("off")
plt.show
```

```python
X = df.iloc[:,1:3001]
X
```

```python
Y = df.iloc[:,-1]
Y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.25,random_state=50)
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train,y_train)
```

```python
knn_pred = knn.predict(x_test)
knn_pred
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
print(classification_report(y_test,knn_pred))
print(sns.heatmap(confusion_matrix(y_test,knn_pred),annot=True,fmt="0.2f"))
print("Accuracy - ",accuracy_score(y_test,knn_pred))
```

```python
from sklearn.svm import SVC
svc = SVC(C=1.0, kernel='rbf',gamma="auto")
svc.fit(x_train,y_train)
```

```python
svc_pred = svc.predict(x_test)
```

```python
print(classification_report(y_test,svc_pred))
print(sns.heatmap(confusion_matrix(y_test,svc_pred),annot=True,fmt="0.2f"))
print("Accuracy - ",accuracy_score(y_test,svc_pred))
```
